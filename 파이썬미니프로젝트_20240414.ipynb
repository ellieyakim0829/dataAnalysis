```python
# 주소록 프로그램 만들기

menu = ''' ---------[선택메뉴]--------
1. 연락처 입력
2. 연락처 검색
3. 연락처 삭제
4. 연락처 보기
0. 종료하기
'''

address_book = {}  # 주소록을 저장할 딕셔너리

# 함수 선언부분
def book_add():
    name = input('이름:')
    phone_number = input('전화번호:')
    email  = input('E-mail:')
    addr =input('Address:')

    # 주소록에 딕셔너리 형태로 저장 {이름 key, value 값으로 전화번호, 이메일, 주소를 딕셔너리에 저장}
    address_book[name] = {'전화번호': phone_number,'E-mail': email,'Address':addr}
    print(f"'{name}'연락처가 등록되었습니다.")

def book_search():   # 연락처 검색
    name_search = input('연락처 검색: ')
    if name_search in address_book:                 # 현재 등록된 딕셔너리에  이름(key)가 있는지 검색
        contact_info = address_book[name_search]    # 해당 이름에 대한 연락처(value) 정보 가져오기
        print(f"이름: {name_search}")                # 주소록 정보 출력
        print(f"전화번호: {contact_info['전화번호']}")
        print(f"E-mail: {contact_info['E-mail']}")
        print(f"Address: {contact_info['Address']}")
    else:
        print("연락처가 등록되어 있지 않습니다..")
        input_name = input('새로운 연락처를 등록하시겠습니니까?(등록:Y, 미등록:N):')  # 미등록된 연락처 추가할지 여부 확인
        if input_name =='Y' or 'y':
            book_add()  # 연락처 추가 함수 호출
        else:
            print(f'등록하지 않습니다.')

def book_delete():  # 연락처 삭제
    name_Delete = input('삭제할 연락처 :')     # 삭제할 연락처 이름 받기
    if name_Delete in address_book:           # 삭제할 연락처가 딕셔너리에 있는지 검색
        del address_book[name_Delete]         # 딕셔너리 주소록에서 연락처 삭제
        print(f"연락처가 삭제 되었습니다: {name_Delete}")
    else:
        print(f"등록되지 않은 이름입니다..")

def book_show():
    for name, contact_info in address_book.items():   # 주소록의 모든 연락처에 대해 반복
        print(f"이름: {name}")                        # 현재 등록된 모든 {이름 key, value 값으로 전화번호, 이메일, 주소} 출력
        print(f"전화번호: {contact_info['전화번호']}")
        print(f"E-mail: {contact_info['E-mail']}")
        print(f"Address: {contact_info['Address']}")


# 메인코드 부분
while True:
    print(menu)
    select = input('-->메뉴 선택: ')
    if select == '0':    # 프로그램 종료
        break
    elif select == '1':  # 연락처 입력
        book_add()       # 연락처 입력 함수 호출
    elif select == '2':  # 연락처 검색
        book_search()    # 연락처 검색 함수 호출
    elif select == '3':  # 연락처 삭제
        book_delete()    # 연락처 삭제 함수 호출
    elif select == '4':  # 연락처 보기
        book_show()      # 연락처 보기 함수 호출

print('프로그램을 종료합니다.')
```

```python
# 함수사용 영어단어장(단어관리)프로그램 만들기

menu = ''' ---------[선택메뉴]--------
1. 단어 등록
2. 단어 검색
3. 단어 삭제
4. 모든 단어 보기
0. 종료하기
'''

English_dict = {}

# 함수 선언부분

def w_add():
    word = input('새단어 :')           # 새로 등록할 단어(key) 입력
    word_mean =input(f'{word}의 뜻 :') # value 값 입력
    English_dict[word ] = word_mean    # 딕셔너리에서 {Key,Value} 등록
    print(f"'{word}'가 등록되었습니다.")

def w_search():
    word_search = input('검색할 단어: ')        # 검색한 단어(key) 입력
    if word_search in English_dict:             # 현재 등록된 딕셔너리에  단어(key)가 있는지 검색
        word_mean = English_dict[word_search]   # 딕셔너리에서 Key를 사용해 Value 얻기
        print(f"{word_search}에 해당하는 값은 {word_mean}입니다.")
    else:
        print("미등록 단어입니다.")
        input_word = input('새로운 단어를 등록하시겠습니니까?(등록:Y, 미등록:N):') # 미등록 단어(key)인 경우 단어 추가여부 확인하기
        if input_word =='Y' or 'y':
            w_add()  # 등록할 단어 함수 호출
        else:
            print(f'등록하지 않습니다.')
def w_delete():
    word_Delete = input('삭제할 단어 :') # 삭제할 단어(key) 입력
    if word_Delete in English_dict:      # 현재 등록된 딕셔너리에  단어(key)가 있는지 검색
        del English_dict[word_Delete]    # key값 삭제
        print(f"단어가 삭제 되었습니다: {word_Delete}")
    else:
        print(f"미등록 단어입니다..")

def w_show():
    for word, word_mean in English_dict.items():
        print(f"{word}: {word_mean}")    # 현재 등록된 모든 단어(key)와 뜻(value) 출력

# 메인코드 부분
while True:
    print(menu)
    select = input('-->메뉴 선택: ')
    if select == '0':   # 프로그램 종료
        break
    elif select == '1': # 단어 등록
        w_add()         # 함수 호출
    elif select == '2': # 단어 검색
        w_search()      # 함수 호출
    elif select == '3': # 단어 삭제
        w_delete()      # 함수 호출
    elif select == '4': # 현황보기
        w_show()        # 함수 호출

print('프로그램을 종료합니다.')
```
